```python
import os
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear')
```

```python
current_folder = globals()['_dh'][0]
CSV_FILE = os.path.join(current_folder, '..', '..', 'csv', 'organizations-100.csv')

df = pd.read_csv(CSV_FILE)
```

```python
df.keys()
```

```python
x = df[['Founded', 'Number of employees']].values
y = df['Country'].values
model.max_iter = 1000000
model.fit(x, y)
```

```python
print(model.coef_, model.intercept_)
```

```python
df.head(1)
```

```python
# Predict individual samples
model.predict([[1990, 3498]])
```

```python
# Predict for all
y_pred = model.predict(x)
y_pred
```

```python
# Score the accuracy of the model
print((y == y_pred).sum() / y.shape[0])
# Use built-in
print(model.score(x, y))
```
